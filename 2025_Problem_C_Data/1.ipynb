```python
import pandas as pd
import matplotlib.pyplot as plt
## import seaborn as sns
import numpy as np
```

```python
athletes = pd.read_csv('data/summerOly_athletes.csv')
athletes
```

```python
athletes['Sport_Split'] = athletes['Sport'].str.split(', ')
athletes['Sport_Count'] = athletes['Sport_Split'].apply(len)
multiple_sport_rows = athletes[athletes['Sport_Count'] > 1]
single_sport_rows = athletes[athletes['Sport_Count'] == 1]
multiple_sport_rows
```

```python
sport_event_dict = single_sport_rows.groupby('Sport')['Event'].apply(set).to_dict()
sport_event_dict['Swimming']
# athletes[athletes['Sport_Count'] > 1]
```

```python
def clean_sport(row):
    
    sport_list = row['Sport_Split']
    event = row['Event']
    valid_sports = []
    
    for sport in sport_list:
        # print(sport)
        if sport in sport_event_dict and event in sport_event_dict[sport]:
            valid_sports.append(sport)
    
    if len(valid_sports) == len(sport_list):
        valid_sports = sport_list
    
    row['Sport'] = ', '.join(valid_sports)
    row['Sport_Split'] = valid_sports
    row['Sport_Count'] = len(valid_sports)
    
    return row

# 对 multiple_sport_rows 应用清理函数
# cleaned_rows = multiple_sport_rows.apply(clean_sport, axis=1)                                

# 将清理后的行更新回原数据
# athletes.update(cleaned_rows)

# 删除 Sport_Count 为 0 的行（即没有合法 Sport 的行）
athletes = athletes[athletes['Sport_Count'] > 0]

# 输出清理后的数据
athletes[athletes['Sport_Count'] > 1]
```
