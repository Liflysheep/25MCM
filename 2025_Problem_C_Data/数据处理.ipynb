```python
import pandas as pd

# 1. 处理缺失值：将年份列中非整数的值替换为 NaN
# 提取年份列（从1896到2024）
year_columns = data_programs.columns[4:]  # 从第5列开始是年份列

# 将非整数的值替换为 NaN
for col in year_columns:
    data_programs[col] = pd.to_numeric(data_programs[col], errors='coerce')

# 2. 删除包含 "Skating" 和 "Ice Hockey" 的行
# 假设 "Sport" 列包含运动项目的名称
data_programs = data_programs[~data_programs['Sport'].isin(['Skating', 'Ice Hockey'])]

# 3. 使用最邻近插值填补缺失值
# 对每行（每个运动项目）的时间序列数据进行最邻近插值
data_programs[year_columns] = data_programs[year_columns].interpolate(method='nearest', axis=0, limit_direction='both')

# 4. 确保填补值为非负整数
data_programs[year_columns] = data_programs[year_columns].round().astype(int)
data_programs[year_columns] = data_programs[year_columns].clip(lower=0)

# 查看处理后的数据和 NaN 值统计
print("处理后的数据：")
print(data_programs.head())

# 检查是否还有缺失值
print("\n每列中 NaN 值的数量：")
print(data_programs.isna().sum())
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error
import pandas as pd  # 确保导入 pandas


# 假设 data_programs 是已经处理过的数据
# 2. 删除包含 "Skating" 和 "Ice Hockey" 的行
# 假设 "Sport" 列包含运动项目的名称
data_programs_test = data_programs[~data_programs['Sport'].isin(['Total events', 'Total disciplines','Total sports'])]
# 1. 模拟缺失值
# 随机选择 10% 的数据作为模拟缺失值
np.random.seed(42)  # 设置随机种子以确保可重复性
mask = np.random.rand(*data_programs_test[year_columns].shape) < 0.10  # 10% 的数据作为模拟缺失值
simulated_missing = data_programs_test[year_columns].copy()
simulated_missing[mask] = np.nan  # 将选中的数据设置为 NaN

# 2. 使用最邻近插值填补模拟缺失值
filled_data = simulated_missing.interpolate(method='nearest', axis=0, limit_direction='both')

# 3. 确保填补值为非负整数
# 检查是否存在 NaN 或 inf，如果有则填充为 0
filled_data = filled_data.fillna(0)  # 将 NaN 替换为 0
filled_data = np.nan_to_num(filled_data, nan=0, posinf=0, neginf=0)  # 处理 inf
filled_data = filled_data.round().astype(int)  # 取整并转换为整数
filled_data = filled_data.clip(min=0)  # 确保为非负数

# 将 filled_data 转换回 Pandas DataFrame
filled_data = pd.DataFrame(filled_data, columns=year_columns)  # 转换为 DataFrame 并指定列名

# 4. 计算误差指标
# 提取真实值和填补值
true_values = data_programs_test[year_columns].values[mask]
imputed_values = filled_data.values[mask]  # 使用 .values 提取 NumPy 数组

# 计算均方误差 (MSE) 和平均绝对误差 (MAE)
mse = mean_squared_error(true_values, imputed_values)
mae = mean_absolute_error(true_values, imputed_values)

print(f"均方误差 (MSE): {mse}")
print(f"平均绝对误差 (MAE): {mae}")

# 5. 可视化对比
# 随机选择一列进行可视化
sample_col = np.random.choice(year_columns)  # 随机选择一个年份列
sample_true = data_programs_test[sample_col]
sample_imputed = filled_data[sample_col]  # 使用列名直接提取数据

# 创建画布
plt.figure(figsize=(12, 7), dpi=100)  # 提高分辨率和调整尺寸

# 绘制曲线
plt.plot(sample_true, 
         label='True Values', 
         marker='o', 
         markersize=8, 
         linewidth=2,
         color='#2c7bb6',  # 蓝色系
         alpha=0.8)

plt.plot(sample_imputed, 
         label='Imputed Values', 
         marker='x', 
         markersize=10,
         linewidth=2, 
         linestyle='--', 
         color='#d7191c',  # 红色系
         alpha=0.8)

# 美化设置
plt.title(f"Comparison of True Values and Imputed Values ({sample_col})", 
          fontsize=14, 
          fontweight='bold',
          pad=20)  # 标题下移

plt.xlabel("Row Index", 
           fontsize=12, 
           labelpad=10)  # X轴标签
plt.ylabel("Value", 
           fontsize=12, 
           labelpad=10)  # Y轴标签

plt.grid(True, 
         linestyle='--', 
         alpha=0.6)  # 添加虚线网格
plt.xticks(fontsize=10)  # 调整刻度字体
plt.yticks(fontsize=10)

# 添加图例
legend = plt.legend(fontsize=12, 
                    frameon=True, 
                    shadow=True,
                    edgecolor='#444444',
                    bbox_to_anchor=(1.02, 1),  # 将图例移到图外右侧
                    loc='upper left')
legend.get_frame().set_facecolor('#f5f5f5')  # 设置图例背景色

# 优化布局
plt.tight_layout()  # 自动调整元素间距
plt.gca().set_facecolor('#f9f9f9')  # 设置图表背景色

# 显示图表
plt.show()
```

```python
import seaborn as sns

# 计算关键统计量对比
original_stats = data_programs[year_columns].describe()
filled_stats = data_programs[year_columns].describe()

# 可视化分布差异
plt.figure(figsize=(12, 6))
sns.kdeplot(
    data_programs[year_columns].values.flatten(), 
    label='Original Data', 
    fill=True, 
    alpha=0.3
)
sns.kdeplot(
    data_programs[year_columns].values.flatten(), 
    label='Imputed Data', 
    fill=True, 
    alpha=0.3
)
plt.title("Distribution Comparison")
plt.legend()
plt.show()

# 打印统计量差异
print("统计量差异对比:")
print((filled_stats - original_stats).loc[['mean', 'std']])
```
